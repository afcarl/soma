# 6 Searching for Structure in Point Data

In this chapter we deal with 3 problems related with "exploratory data analysis". Namely, given a sample of N points in D-dimensional space we have the following questions:

1. What would be the pdf from which our sample has been drawn? - Density estimation
2. Do our datapoints randomly distributed, or can we identify distinct groups? - Cluster finding
3. If indeed there are groupings (structure) in our data, how can we characterize them? - Statistical description of observed structure

## 6.1 Nonparametric Density Estimation

Here we are not trying to fit the data to any particular model, we are just interested to estimate the pdf (numerically) so that we can analyze it, sample it...

### 6.1.1 Kernel Density Estimation (KDE)

Finding overdensities in the data is not trivial. While with 1-D or 2-D data we could rely on our eyes to find overdensities (although a 1:1 map is rarely of any use), for data of higher dimensionality visual inspection is ruled out.

We have already seen histograms, but they suffer from 2 main shortcommings: the choice of both the bin size and its their position alter the result and can drive us to different interpretations of the same data, as seen in the next figure where bin location is shifted.

```python
Image(url='http://www.astroml.org/_images/fig_hist_to_kernel_1.png')
```

KDE adresses the second problem (bin location), by assigning one bin to each datapoint and letting bins to overlap. That way it is data what drive the position, and it is not any more a parameter of choice. The method allows also to use different shapes for the bins, which we will can from now on "kernels".

Some of the most commonly used shapes for the kernels are:

i) Gaussian: $K(u)=\frac{1}{(2\pi)^{D/2}}e^{-u^2/2}$

ii) Top-hat: $K(u)=\frac{1}{V_D(1)} \; if \;  u\leq 1 \; and \; K(u)=0 \; if \; u>1$

iii) Exponential: $K(u)=\frac{1}{D!V_D(1)}e^{-|u|}$

With $V_D(r)$ the volume of a D-dimensional hypersphere of radius r

Nevertheless, the kernel with the minimum variance of the density estimation (given h) is the Epanechnikov kernel:

$K(u)=\frac{3}{4}\left(1-u^2\right) \; for \, |u| \leq 1 and \, 0 \, otherwise$

```python
Image(url='http://scikit-learn.org/stable/_images/plot_kde_1d_002.png')
```

All these shapes have a mean of zero and a variance greater than zero. We can estimate the pdf of a set of measurements ${x_i}$ by

$\hat{f}_N(x)=\frac{1}{Nh^D}\sum\limits_{i=1}^N K \left( \frac{d(x,x_i)}{h} \right)$

There is still unresolved the question of how big (h) te kernels should be. There is a trade-off between too much locality (with noisy estimations) and too much smoothness (with loss of information), as we can see in the next figure

```python
Image(url='http://www.astroml.org/_images/fig_hist_to_kernel_2.png')
```

We want to find the size of the kernel that minimizes the cost of out-of-sample data. The book offers two potential cost functions to optimize:

The first assumes that the cost of leaving one datapoint out is the likelihood, thus the total cost is the sum

$CV_l(h)=\frac{1}{N}\sum\limits_{i=1}^N log \, \hat{f}_{h,-i}(x_i)$

The second score comes from using the mean integrated square error (MISE)

$CV_{L_2}=\int \hat{f}^2_h - 2 \frac{1}{N}\sum\limits_{i=1}^N \hat{f}_{h,-i}(x_i)$

The error of KDE using the optimized h converges like $O(N^{-4/5})$, and it can be proved that no other density estimator converges faster

###### Implementing KDE

We are going to use the class KernelDensity from sklearn.neighbors. We initialize it with just 2 inputs, the bandwidth h and the type of kernel. Once initialize, we can apply among others the following methods:

fit(X) : fits the model on the data

score_samples(X) : evaluates the model on the data, yields an array of log(density) evaluations


And we are going to apply it to the book example, galaxy distribution in the "Great Wall" using data from SDSS

```python
# The usual imports
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

# We import our data: galaxy distribution
from astroML.datasets import fetch_great_wall
X = fetch_great_wall()

# And we plot them to have an idea of the distribution
fig0 = plt.figure(figsize=(10,5))
ax0 = plt.axes()
ax0.scatter(X[:, 1], X[:, 0], s=1, lw=0, c='k')
ax0.set_xlim(-300, 200)
ax0.set_ylim(-375, -175)
ax0.set_xlabel('$y$ (Mpc)')
ax0.set_ylabel('$x$ (Mpc)')
plt.show()
```

```python
# Imports to implement KDE
from scipy.spatial import cKDTree
from scipy.stats import gaussian_kde
from sklearn.neighbors import KernelDensity
from matplotlib.colors import LogNorm
# To make the figure look good like in the textbook
from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=16, usetex=True)
#------------------------------------------------------------
# Create  the grid on which to evaluate the results
Nx = 50
Ny = 125
xmin, xmax = (-375, -175)
ymin, ymax = (-300, 200)
#------------------------------------------------------------
# Evaluate for several models
Xgrid = np.vstack(map(np.ravel, np.meshgrid(np.linspace(xmin, xmax, Nx),
                                            np.linspace(ymin, ymax, Ny)))).T

kernels = ['gaussian', 'tophat', 'exponential', 'epanechnikov']
dens = []
hgauss = 5.0
htophat = 5.0
hexp = 5.0
hepan = 5.0

kde1 = KernelDensity(hgauss, kernel='gaussian')    # Creates a gaussian kde model
log_dens1 = kde1.fit(X).score_samples(Xgrid)    # Fits the model to the data, yields log(density)
dens1 = X.shape[0] * np.exp(log_dens1).reshape((Ny, Nx))    # Reshapes the result and normalizes with number of galaxies

kde2 = KernelDensity(htophat, kernel='tophat')
log_dens2 = kde2.fit(X).score_samples(Xgrid)
dens2 = X.shape[0] * np.exp(log_dens2).reshape((Ny, Nx))

kde3 = KernelDensity(hexp, kernel='exponential')
log_dens3 = kde3.fit(X).score_samples(Xgrid)
dens3 = X.shape[0] * np.exp(log_dens3).reshape((Ny, Nx))

kde4 = KernelDensity(hepan, kernel='epanechnikov')
log_dens4 = kde4.fit(X).score_samples(Xgrid)
dens4 = X.shape[0] * np.exp(log_dens4).reshape((Ny, Nx))
#------------------------------------------------------------
# Plot the results
fig = plt.figure(figsize=(20, 10))
fig.subplots_adjust(left=0.12, right=0.95, bottom=0.2, top=0.9,
                    hspace=0.01, wspace=0.01)

# First plot: gaussian kernel
ax1 = plt.subplot(221, aspect='equal')
ax1.imshow(dens1.T, origin='lower', norm=LogNorm(),
           extent=(ymin, ymax, xmin, xmax), cmap=plt.cm.binary)
ax1.text(0.95, 0.9, "Gaussian $(h=5)$", ha='right', va='top',
         transform=ax1.transAxes,
         bbox=dict(boxstyle='round', ec='k', fc='w'))

# Second plot: top-hat kernel
ax2 = plt.subplot(222, aspect='equal')
ax2.imshow(dens2.T, origin='lower', norm=LogNorm(),
           extent=(ymin, ymax, xmin, xmax), cmap=plt.cm.binary)
ax2.text(0.95, 0.9, "top-hat $(h=5)$", ha='right', va='top',
         transform=ax2.transAxes,
         bbox=dict(boxstyle='round', ec='k', fc='w'))
ax2.images[0].set_clim(0.01, 0.8)

# Third plot: exponential kernel
ax3 = plt.subplot(223, aspect='equal')
ax3.imshow(dens3.T, origin='lower', norm=LogNorm(),
           extent=(ymin, ymax, xmin, xmax), cmap=plt.cm.binary)
ax3.text(0.95, 0.9, "exponential $(h=5)$", ha='right', va='top',
         transform=ax3.transAxes,
         bbox=dict(boxstyle='round', ec='k', fc='w'))

# Fourth plot: epanechnikov kernel
ax4 = plt.subplot(224, aspect='equal')
ax4.imshow(dens4.T, origin='lower', norm=LogNorm(),
           extent=(ymin, ymax, xmin, xmax), cmap=plt.cm.binary)
ax4.text(0.95, 0.9, "Epanechnikov $(h=5)$", ha='right', va='top',
         transform=ax4.transAxes,
         bbox=dict(boxstyle='round', ec='k', fc='w'))
ax4.images[0].set_clim(0.01, 0.8)

for ax in [ax1, ax2, ax3, ax4]:
    ax.set_xlim(ymin, ymax - 0.01)
    ax.set_ylim(xmin, xmax)

for ax in [ax1, ax2]:
    ax.xaxis.set_major_formatter(plt.NullFormatter())

for ax in [ax3, ax4]:
    ax.set_xlabel('$y$ (Mpc)')

for ax in [ax2, ax4]:
    ax.yaxis.set_major_formatter(plt.NullFormatter())

for ax in [ax1, ax3]:
    ax.set_ylabel('$x$ (Mpc)')

plt.show()
```

### 6.1.2. (6.2) Nearest-Neighbor Density Estimation

A potential problem with KDE is that it uses the same bandwidth throrough the dataset. Areas that are sparsely populated (e.g. distribution tails) may be estimated noisily. Using a different bandwidth could potentially address that issue.

An alternative to the KDE method is just to assume constant density locally. That density can be estimated as just the nearest K neighbours divided by the volude define by the furthest one:
  
  $\hat{f}_K(x)=\frac{K}{V_D(d_K)}$
  
  And, if we don't want to sum how many neighbours are within the volume, we can just normalize ex-post so that the product of the estimated density and volume equals the total number of data points.
  
  Here the density is driven by the hiper-volume (denominator) instead of the numberator. It is as if we were adapting the bandwidth (smaller where point density is higher)

Apart from this "simple" estimator, it can be proven that errors improve if we use a "bayesian" estimator that incorporates distances to all neighbors:

$\hat{f}_K(x)=\frac{C}{\sum\limits_{i=1}^N d_i^D}, \; with \; C=\frac{K(K+1)}{2V_D(1)}$

###### Implementing nearest neighbor

We are goint to use the class KNeighborsDensity from astroML. We only need to define 2 parameters:

method : 'simple' if we want to use the distance to the k-th neighbor or 'bayesian' if we want to use the 'refined' version of it (using all distances until the k-th neighbor)

n_neighbors : number of neighbors to consider (k)

```python
# Imports to implement nearest neighbors and kde algorithms
from astroML.density_estimation import KNeighborsDensity

knn5 = KNeighborsDensity('bayesian', 5)
dens_k5 = knn5.fit(X).eval(Xgrid).reshape((Ny, Nx))

knn40 = KNeighborsDensity('bayesian', 40)
dens_k40 = knn40.fit(X).eval(Xgrid).reshape((Ny, Nx))

knn40s = KNeighborsDensity('simple', 40)
dens_k40s = knn40s.fit(X).eval(Xgrid).reshape((Ny, Nx))

#------------------------------------------------------------
# Plot the results
fig = plt.figure(figsize=(20, 10))
fig.subplots_adjust(left=0.12, right=0.95, bottom=0.2, top=0.9,
                    hspace=0.01, wspace=0.01)

# First plot: KDE (gaussian)
ax1 = plt.subplot(221, aspect='equal')
ax1.imshow(dens1.T, origin='lower', norm=LogNorm(),
           extent=(ymin, ymax, xmin, xmax), cmap=plt.cm.binary)
ax1.text(0.95, 0.9, "Gaussian $(h=5)$", ha='right', va='top',
         transform=ax1.transAxes,
         bbox=dict(boxstyle='round', ec='k', fc='w'))

# Second plot: KNN, k=5
ax2 = plt.subplot(222, aspect='equal')
ax2.imshow(dens_k5.T, origin='lower', norm=LogNorm(),
           extent=(ymin, ymax, xmin, xmax), cmap=plt.cm.binary)
ax2.text(0.95, 0.9, "$k$-neighbors $(k=5)$ (Bayesian)", ha='right', va='top',
         transform=ax2.transAxes,
         bbox=dict(boxstyle='round', ec='k', fc='w'))

# Third plot: KNN, k=40
ax3 = plt.subplot(223, aspect='equal')
ax3.imshow(dens_k40.T, origin='lower', norm=LogNorm(),
           extent=(ymin, ymax, xmin, xmax), cmap=plt.cm.binary)
ax3.text(0.95, 0.9, "$k$-neighbors $(k=40)$ (Bayesian)", ha='right', va='top',
         transform=ax3.transAxes,
         bbox=dict(boxstyle='round', ec='k', fc='w'))

# Fourth plot: KNN, k=40 (simple)
ax4 = plt.subplot(224, aspect='equal')
ax4.imshow(dens_k40s.T, origin='lower', norm=LogNorm(),
           extent=(ymin, ymax, xmin, xmax), cmap=plt.cm.binary)
ax4.text(0.95, 0.9, "$k$-neighbors $(k=40)$ (Simple)", ha='right', va='top',
         transform=ax4.transAxes,
         bbox=dict(boxstyle='round', ec='k', fc='w'))

for ax in [ax1, ax2, ax3, ax4]:
    ax.set_xlim(ymin, ymax - 0.01)
    ax.set_ylim(xmin, xmax)

for ax in [ax1, ax2]:
    ax.xaxis.set_major_formatter(plt.NullFormatter())

for ax in [ax3, ax4]:
    ax.set_xlabel('$y$ (Mpc)')

for ax in [ax2, ax4]:
    ax.yaxis.set_major_formatter(plt.NullFormatter())

for ax in [ax1, ax3]:
    ax.set_ylabel('$x$ (Mpc)')

plt.show()
```

### 6.1.3 (6.1.2) Non parametric density estimation with errors

In the book this point is treated wihtin KDE, but could / could not be this method applied to kNN??

Extrapolating from the case of a 1-D problem with homoscedastic errors, if we know the error distribution (or have an estimate) the procedure would be:

1. Obtain the pdf of the observed data with errors, $f(x)$
2. Compute its Fourier transform (numerically), $\mathcal{F}(k)$
3. Assume (or model) a noise distribution $g(x)$
4. Compute its Fourier transform (numerically), $\mathcal{G}(k)$
5. The Fourier transform of the true (noise free) distribution is $\mathcal{H}(k) = \mathcal{F}(k)/\mathcal{G}(k)$
6. Compute the inverse Fourier transform of $\mathcal{H}(k)$ to get $h(x)$

## 6.3 Parametric Density Estimation

In some occasions, we may be interested to see how the distribution of our data adjust to a (small) set of distributions. We can do that fitting for kernel location as well. If we pick many distributions, it will tend to KDE, while if we just pick a few, it will help cluster the data. 

### 6.3.1 Gaussian Mixture Model (GMM)

Let's try to fit our data's pdf with a sum of gaussians:

$\rho(\mathbf{x})=Np(\mathbf{x})=N\sum\limits_{j=1}^M\alpha_j\mathcal{N}\left(\mu_j,\Sigma_j\right)$

Once fitted a number of gaussians, we can evaluate the goodness of the fit evaluating its AIC (frequentist) or BIC (bayesian) score, or both. By repeating the process for different values of N, we can find the optimal fit. As a reminder:

$AIC \equiv -2ln \left(L^0(M)\right) + 2k + \frac{2k(k+1)}{N-k-1}$

$BIC \equiv -2ln \left(L^0(M)\right) + klnN$

We should not confuse the N gaussians obtained through this procedure as the number of clusters in our data (for instance they may not follow gaussian distributions in the first place!). There are better methods for clustering

###### Implementing GMM

```python
# Imports to implement GMM
from sklearn.mixture import GMM
from astroML.decorators import pickle_results
#------------------------------------------------------------
# Create a function which will save the results to a pickle file
#  for large number of clusters, computation will take a long time!
@pickle_results('great_wall_GMM.pkl')
def compute_GMM(n_clusters, n_iter=1000, min_covar=3, covariance_type='full'):
    clf = GMM(n_clusters, covariance_type=covariance_type,
              n_iter=n_iter, min_covar=min_covar)
    clf.fit(X)
    print "converged:", clf.converged_
    return clf

# We do not upload again the data, nor recreate the grid to evaluate the density estimator
#------------------------------------------------------------
# Compute the results
#
# we'll use 100 clusters.  In practice, one should cross-validate
# with AIC and BIC to settle on the correct number of clusters.
clf = compute_GMM(n_clusters=100)
log_dens = clf.score(Xgrid).reshape(Ny, Nx)
```

### 6.3.3 Extreme Deconvolution (XD): GMM with errors

As in GMM, we assume that our data come from a set of gaussians:

$\rho(\mathbf{x})=Np(\mathbf{x})=N\sum\limits_{j=1}^M\alpha_j\mathcal{N}\left(\mu_j,\Sigma_j\right)$

but in this case, out observed data $\mathbf{x_i}$ have noise and are thus related to the "true" values by

$\mathbf{x_i}=\mathbf{R_ix_i}+\epsilon_i$

where $\mathbf{R_i}$ is the projection matrix (may not be invertible). We further assume that errors follow a gaussian distribution with mean zero and variance $\mathbf{S_i}$.

Out problem can be stated by, given $\mathbf{R_i}$ and $\mathbf{S_i}$, find the parameters that define our model pdf, {$\alpha_i,\;\mu_i,\;\Sigma_i$}

The book describes an EM (expectation maximization) iterative algorithm (we need to start with some initial guesses for our parameters) which we are not going to describe here in detail. It is based on the fact that what truly complicates the optimization process is that we don't know to which gaussian each datapoint comes from. If we knew, the problem would reduce to just fit a set of gaussians to a set of subsamples, and that is much easier.

We assume an initial probability for each point to come from each one of the model's gaussians (E step), compute the parameters that best fit that assumption (M step) and then re-estimate the initial probabilities (E step) and continue until we achieve convergence.

###### Implementing XD

astroML has a XD implementation class:

astroML.density_estimation.XDGMM(n_components, n_iter=100, tol=1e-5, verbose=False)

from which we will use the following methods:

fit(X) : fits the model obtained to the data

sample([size]) : creates a sample from the model's pdf

We are going to reproduce here the book example. To do so we first do some imports and data work to prepare the samples on which to apply XD

```python
# Imports to implement XD using astroML
from astroML.density_estimation import XDGMM
from astroML.crossmatch import crossmatch
# Imports additional functions used
from astroML.plotting.tools import draw_ellipse
from astroML.stats import sigmaG
# Imports stellar data to apply
from astroML.datasets import fetch_sdss_S82standards, fetch_imaging_sample
#------------------------------------------------------------
# define u-g-r-i-z extinction from Berry et al, arXiv 1111.4985
# multiply extinction by A_r
extinction_vector = np.array([1.810, 1.400, 1.0, 0.759, 0.561])
#----------------------------------------------------------------------
# Fetch and process the noisy imaging data
data_noisy = fetch_imaging_sample()
# select only stars
data_noisy = data_noisy[data_noisy['type'] == 6]
# Get the extinction-corrected magnitudes for each band
X = np.vstack([data_noisy[f + 'RawPSF'] for f in 'ugriz']).T
Xerr = np.vstack([data_noisy[f + 'psfErr'] for f in 'ugriz']).T
# extinction terms from Berry et al, arXiv 1111.4985
X -= (extinction_vector * data_noisy['rExtSFD'][:, None])
#----------------------------------------------------------------------
# Fetch and process the stacked imaging data
data_stacked = fetch_sdss_S82standards()
# cut to RA, DEC range of imaging sample
RA = data_stacked['RA']
DEC = data_stacked['DEC']
data_stacked = data_stacked[(RA > 0) & (RA < 10) &
                            (DEC > -1) & (DEC < 1)]
# get stacked magnitudes for each band
Y = np.vstack([data_stacked['mmu_' + f] for f in 'ugriz']).T
Yerr = np.vstack([data_stacked['msig_' + f] for f in 'ugriz']).T
# extinction terms from Berry et al, arXiv 1111.4985
Y -= (extinction_vector * data_stacked['A_r'][:, None])
# quality cuts
g = Y[:, 1]
mask = ((Yerr.max(1) < 0.05) &
        (g < 20))
data_stacked = data_stacked[mask]
Y = Y[mask]
Yerr = Yerr[mask]
#----------------------------------------------------------------------
# cross-match
#  the imaging sample contains both standard and variable stars.  We'll
#  perform a cross-match with the standard star catalog and choose objects
#  which are common to both.
Xlocs = np.hstack((data_noisy['ra'][:, np.newaxis],
                   data_noisy['dec'][:, np.newaxis]))
Ylocs = np.hstack((data_stacked['RA'][:, np.newaxis],
                   data_stacked['DEC'][:, np.newaxis]))
print "number of noisy points:  ", Xlocs.shape
print "number of stacked points:", Ylocs.shape
# find all points within 0.9 arcsec.  This cutoff was selected
# by plotting a histogram of the log(distances).
dist, ind = crossmatch(Xlocs, Ylocs, max_distance=0.9 / 3600.)

noisy_mask = (~np.isinf(dist))
stacked_mask = ind[noisy_mask]

# select the data
data_noisy = data_noisy[noisy_mask]
X = X[noisy_mask]
Xerr = Xerr[noisy_mask]

data_stacked = data_stacked[stacked_mask]
Y = Y[stacked_mask]
Yerr = Yerr[stacked_mask]

# double-check that our cross-match succeeded
assert X.shape == Y.shape
print "size after crossmatch:", X.shape
```

We can now use the XD implementation in astroML

```python
# perform extreme deconvolution on the noisy sample
# first define mixing matrix W
W = np.array([[0, 1, 0, 0, 0],    # g magnitude
              [1, -1, 0, 0, 0],   # u-g color
              [0, 1, -1, 0, 0],   # g-r color
              [0, 0, 1, -1, 0],   # r-i color
              [0, 0, 0, 1, -1]])  # i-z color

X = np.dot(X, W.T)
Y = np.dot(Y, W.T)

# compute error covariance from mixing matrix
Xcov = np.zeros(Xerr.shape + Xerr.shape[-1:])
Xcov[:, range(Xerr.shape[1]), range(Xerr.shape[1])] = Xerr ** 2

# each covariance C = WCW^T
# best way to do this is with a tensor dot-product
Xcov = np.tensordot(np.dot(Xcov, W.T), W, (-2, -1))

#----------------------------------------------------------------------
# This is a long calculation: save results to file
@pickle_results("XD_stellar.pkl")
def compute_XD(n_clusters=12, rseed=0, n_iter=100, verbose=True):
    np.random.seed(rseed)
    clf = XDGMM(n_clusters, n_iter=n_iter, tol=1E-5, verbose=verbose)
    clf.fit(X, Xcov)
    return clf

clf3 = compute_XD(12)

#------------------------------------------------------------
# Fit and sample from the underlying distribution
np.random.seed(42)
X_sample = clf3.sample(X.shape[0])
```
