Now, let's do something a little more interesting. Let's make a simple color-magnitude diagram. We have to use Matplotlib now.

```python
%pylab inline
from matplotlib import pyplot as plt

stars = data[data['type'] == 6]

plt.plot(stars['gRaw'][:400]-stars['rRaw'][:400],stars['gRaw'][:400],'r.')

xlim(-1,3.5)
ylim(26,14)

xlabel("g-r")
ylabel("g")

plt.show()
```

Maybe there's a bit of a main sequence there. It's a bit hard to tell because these are all field stars. 

Now let's make a color-color diagram.

```python
plt.plot(stars['gRaw']-stars['rRaw'],stars['rRaw']-stars['iRaw'],'r.',alpha=0.01)
plt.show()
```

But clearly these axes are wrong. When we zoom in:

```python
plt.plot(stars['gRaw']-stars['rRaw'],stars['rRaw']-stars['iRaw'],'r.',alpha=1)

xlim(-1,3)
ylim(-1,2.5)

plt.show()
```

This is cluttered! Let's fix with a histogram.

```python
import numpy as np
from astroML.plotting import scatter_contour

g = stars["gRaw"]
r = stars["rRaw"]
i = stars["iRaw"]

fig, ax = plt.subplots(figsize=(10,7.5))

scatter_contour(g - r, r - i, threshold=200, log_counts=True, ax=ax,
                histogram2d_args=dict(bins=40),
                plot_args=dict(marker=',', linestyle='none', color='black'),
                contour_args=dict(cmap=plt.cm.bone))


ax.set_xlim(-0.6, 2.5)
ax.set_ylim(-0.6, 2.5)

plt.show()
```

But clearly there is a problem here. Why? The binning is too coarse. Let's see if we can fix this. What happens when we change the number of bins?

```python

fig, ax = plt.subplots(figsize=(10,7.5))

scatter_contour(g - r, r - i, threshold=200, log_counts=True, ax=ax,
                histogram2d_args=dict(bins=np.linspace(-0.5,2.5,30)),
                plot_args=dict(marker=',', linestyle='none', color='black'),
                contour_args=dict(cmap=plt.cm.bone))


ax.set_xlim(-0.6, 2.5)
ax.set_ylim(-0.6, 2.5)

plt.show()
```

What is the "correct" number of bins? It depends on the data, and the structure of the data. As a rule of thumb, for $N$ data points, use $\sqrt{N}$ bins (at least in your region of interest. 

For a 2D histogram, this quickly becomes difficult: for $N$ data points, use $N^{1/4}$ bins on a side.



```python
g.shape
```

Since there are $\sim$82000 data points, you might want to try a grid with 17x17 bins. But as we can see this region is mostly white space, so more bins are justified. You'll have to experiment with it to get the correct size. Too many bins and unphysical structure will appear in your data.

Let's try one more thing now, using astroML data sets. Let's plot the list of solar system objects. The first time you run it, you'll have to download it. It is about 25 MB.

```python
from astroML.datasets import fetch_moving_objects

data = fetch_moving_objects(Parker2008_cuts=True)
data.dtype.names[:]

a = data['aprime']
sini = data['sin_iprime']

fig, ax = plt.subplots(figsize=(10,7.5))
ax.plot(a, sini, '.', markersize=2, color='black')

plt.show()
```
