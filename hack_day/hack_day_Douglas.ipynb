```python
from astropy.io.fits import open as fits_open
from scipy import fftpack
from fitting import linear_fit
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Open the lightcurve file and get the times and (uncorrected?) photometry from the fits file. This is a random star from Kepler Q10, that I could tell by eye had a periodic signal in it. 

```python
hdu = fits_open("kplr100002377-2010174085026_llc.fits")
time = hdu[1].data.field("TIME")
flux = hdu[1].data.field("SAP_FLUX")
flux_err = hdu[1].data.field("SAP_FLUX_ERR")
hdu.close()
```

```python
plt.figure()
plt.plot(time,flux)
plt.xlabel("time (d)")
plt.ylabel("flux")
```

# Fast Fourier Transform

There are gaps in the data that mean we can't do an FFT on the 
full time series. So take only the last segment to actually fit.  Also, there are some NaNs, so replace them with the average of the values on either side.

```python
x,y = np.float64(time)[(time>505)],np.float64(flux)[(time>505)]
bad = np.where(np.isnan(y))
for i in bad:
    y[i] = (y[i-1] + y[i+1])/2.0
```

Very naively de-trending the data by fitting a line then subtracting the line

```python
lin_fit = linear_fit(x,y)
m = lin_fit[0]
b = lin_fit[1]

lin_model = m*x + b
y = y - lin_model
```

Actually do the FFT.
Output has 
0th value is the zero-frequency term, 
[1:n/2+1] has positive frequency terms
[n/2+1:] has negative frequency terms
where n is the length of the output array

```python
y_fft = fftpack.fft(y)
n = len(y_fft)
```

Transform the FFT results into real frequency space (i.e. use the real time spacing), then transform to period space.

```python

day_to_sec = 24*3600.0
delt = 1.0/(len(x)*np.average(np.diff(x))*day_to_sec)

freq = np.zeros(len(y_fft))
freq[1:n/2+1] = np.arange(1,n/2+1) * delt
freq[n/2+1:] = np.arange(1,n/2+1)[::-1] * delt 

period = 1.0/freq/day_to_sec # in days
```

Plot the subset of the data, along with fourier series truncated after different numbers of modes

```python
ax1 = plt.subplot(111)
ax1.plot(x,y,"k-")
ax1.set_xlabel("time (d)")
ax1.set_ylabel("flux (after subtracting linear model)")
for k in [1,3,5,7]:

    y_fft2 = np.copy(y_fft)
    y_fft2[k + 1:-k] = 0

    y_fit = fftpack.ifft(y_fft2).real

    ax1.plot(x,y_fit,lw=2,label=str(k))
ax1.legend(loc=3,borderaxespad=0.0,title="# of modes",ncol=2,handletextpad=0.05,columnspacing=0.25)
```

Plot the power spectrum as a function of period, find its peak, and note it on the plot.

```python
fund_loc = np.argmax(abs(y_fft))
fund_period = period[fund_loc]
```

```python
ax2 = plt.subplot(111)
ax2.plot(period,abs(y_fft))
ax2.set_xlabel("Period (d)")
ax2.set_ylabel("PSD(P)")
ax2.plot(fund_period,abs(y_fft)[fund_loc],'ro')
```

Phase-fold the light-curve and plot the result

```python
phase = x/fund_period - np.asarray((x/fund_period),np.int)

ax3 = plt.subplot(111)
ax3.plot(phase,y,'ko')
ax3.set_xlabel("Phase")
ax3.set_ylabel("Flux (after subtracting linear model)")
```

See how the rest of the light-curve works with this period. Naively de-trend the entire light curve (should probably have de-trended the whole thing to begin with, oops), remove NaNs, and phase-fold on the period determined above.

```python
x2,y2 = np.float64(time),np.float64(flux)
good = np.where(np.isnan(x2)==False)
x2,y2 = x2[good],y2[good]
bad = np.where(np.isnan(y2))
for i in bad:
    y2[i] = (y2[i-1] + y2[i+1])/2.0
```

```python
lin_fit = linear_fit(x2,y2)
m = lin_fit[0]
b = lin_fit[1]
lin_model = m*x2 + b
y2 = y2 - lin_model
```

```python
phase2 = x2/fund_period - np.asarray((x2/fund_period),np.int)

ax4 = plt.subplot(111)
ax4.plot(phase2,y2,'r.')
ax4.set_xlabel("Phase")
ax4.set_ylabel("Flux (after subtracting linear model)")
ax4.set_ylim(-500,400)
```

So it works ok, the sinusoidal signal is still there, but it's fuzzy enough that either my naive linear de-trending didn't work (distinct possibility), or there's a slight period drift over the 95 day quarter.
