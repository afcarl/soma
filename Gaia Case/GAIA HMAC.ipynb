## 1. HMAC in the context of GAIA's astrophysical parameters inference system (Apsis)

As we have seen in the general description of Apsis, in parallel with the supervised learning modules used there is an unsupervised module, OCA (Object Clustering Analysis). It will be fed BP/RP, RVS, parallax & proper motion data, for all sources (in early cycles only the available dimensions) and will try to cluster them in "natural" groups independently of the results from the other modules. The purpose of that is twofold:

1. Identify new object classes, not considered in the supervised methods (treatment of outliers)

2. Help improve training sets (in a similar way to OA)

Nevertheless, even unsupervised methods require some parametric choices, which raises the question on how "natural" the resulting groups that arise are

## 2. HMAC

HMAC clusters data associating each point to a particular point called mode (hence the Mode Association part of the name. Modes are local maxima in the pdf, which is estimated using KDE. Each point crawls "uphill" the estimated density until it meets its node using a EM-style algorithm. The Hierarchical part of the name comes from the fact that this is done for a range of bandwidths (in the KDE) getting to a hierarchical set of modes. The way this algorithm is implemented is the following

1.Estimate the density of the dataset using KDE, a bandwidth h, and a gaussian kernel

2.For each point in the dataset, ascend through the density estimation (MEM):

Given a mixture model $f(x)=\sum_{k=1}^{K} \pi_k f_k(x)$

For r=0,..., until convergence

i) Expectation step: $p_k=\frac{\pi_k f_k \left(x^{(r)} \right)}{f\left(x^{(r)}\right)},\, k=1,...,K$

ii) Maximization step: $x^{(r+1)}=argmax_x\sum_{k=1}^{K}p_klogf_k(x)=...=sum_{k=1}^{K}p_k\mu_k \, (for \, gaussians)$

3.Clustering by Mode Identifications

Start a MEM loop at each point, arriving to a mode, assign each mode, and the points that have ascended to it with the same cluster label

4.Hierarchy

We start with a bandwidth, h, small enough so that almost any point is its own mode. Then, we repeat the method, using the modes instead of the points as starting points with increasing bandwidths, that is:

$Given\,\{h_i\}:\,x_i \rightarrow \mathcal{M}_{h_1}(x_i) \rightarrow \mathcal{M}_{h_2}(\mathcal{M}_{h_1}(x_i))\rightarrow ...$

### 2.1 Enhancing Speed

HMAC is of order $\mathcal{O}(n^2)$. To reduce the computation time, we can use, instead of every data point, a subset to estimate the density. That is, we apply a pre-clustering (with a different method, for instance K-means) so that we pass from n to m points, being m still large compared with the expected number of clusters. Then, we use this "coarse" pdf to find and assing modes.

## 3. Example of applicatio to data

```python
# Data import
from astroML.datasets import fetch_dr7_quasar
from astroML.datasets import fetch_sdss_sspp
quasars = fetch_dr7_quasar()
stars = fetch_sdss_sspp()
```

```python
# Data conditioning
quasars = quasars[::5]
stars = stars[::5]
# We are going to work with 4 dimensions
quasarsUG = quasars['mag_u']-quasars['mag_g']
quasarsGR = quasars['mag_g']-quasars['mag_r']
quasarsRI = quasars['mag_r']-quasars['mag_i']
quasarsIZ = quasars['mag_i']-quasars['mag_z']

starsUG = stars['upsf']-stars['gpsf']
starsGR = stars['gpsf']-stars['rpsf']
starsRI = stars['rpsf']-stars['ipsf']
starsIZ = stars['ipsf']-stars['zpsf']
# Downsampling to run faster
dsamp = 0.1
numQuas = int(np.round(dsamp*len(quasarsUG)))
numStar = int(np.round(dsamp*len(starsUG)))

SquasarsUG = quasarsUG[0:numQuas]
SquasarsGR = quasarsGR[0:numQuas]
SquasarsRI = quasarsRI[0:numQuas]
SquasarsIZ = quasarsIZ[0:numQuas]

SstarsUG = starsUG[0:numStar]
SstarsGR = starsGR[0:numStar]
SstarsRI = starsRI[0:numStar]
SstarsIZ = starsIZ[0:numStar]
```

```python
# Plotting
f, ax = plt.subplots(1,2,sharex=False,sharey=False, figsize=(15,5))
ax[0].scatter(quasarsUG,quasarsGR,s=4,c='b',alpha=0.25, label='$quasars$')
ax[0].scatter(starsUG, starsGR, s=4, c='r', alpha=0.25, label='$stars$')
ax[0].legend(loc='upper left')
ax[0].set_xlim(-0.5,3.0)
ax[0].set_ylim(-0.2,1.4)
ax[0].set_xlabel('$u-g$')
ax[0].set_ylabel('$g-r$')
ax[0].set_title('$Original \, data$')

ax[1].scatter(SquasarsUG,SquasarsGR,s=4,c='b',alpha=0.25, label='$quasars$')
ax[1].scatter(SstarsUG, SstarsGR, s=4, c='r', alpha=0.25, label='$stars$')
ax[1].legend(loc='upper left')
ax[1].set_xlim(-0.5,3.0)
ax[1].set_ylim(-0.2,1.4)
ax[1].set_xlabel('$u-g$')
ax[1].set_ylabel('$g-r$')
ax[1].set_title('$Subsample$')
```

```python
# Merge data to create input .dat file
UG = np.append(quasarsUG, starsUG)
GR = np.append(quasarsGR, starsGR)
RI = np.append(quasarsRI, starsRI)
IZ = np.append(quasarsIZ, starsIZ)

SUG = np.append(SquasarsUG, SstarsUG)
SGR = np.append(SquasarsGR, SstarsGR)
SRI = np.append(SquasarsRI, SstarsRI)
SIZ = np.append(SquasarsIZ, SstarsIZ)
```

###### Comments

In general it is not obvious, a priori, which level of the clustering dendogram will yield the better clustering of the data. Some general principles can be appied:

i) A level is too low if all clusters are too small
ii) A level is too hight if a large percentage (e.g. 95%) of original datapoints is part of one cluster

In this exaple, it seems that levels 3 and 4 yield the best description of the data

We need to choose which set of bandwidths will be used. Apparently, as long as the bandwidth grid is fine enough, the clustering is not very sensitive to the sequence chosen.
