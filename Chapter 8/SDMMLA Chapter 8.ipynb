# Chapter 8. Regression and Model Fitting

## 8.1 The Regression problem

Our aim here is to study the relationship between different variables, more specifically, infer the expectation value of a dependent variable, y,  given a set of independent variables, x. If we have a model for that dependency, we can then write

$E[y|x] = f(x|\theta)$

where $\theta$ is a vector of k parameters. We may be interested either in predicting y for future x values, in finding dependencies (or not dependencies) in the variables, or in the values of the parameters. This problem, which is easier than tryign to infer the full pdf from which the dataset has been drawn, can still be very computationally expensive. In some cases the problem can be simplified, and there are methods specific to each situation. We can think of 3 aspects that determine how complex the problem is:

1. Model: if the model we think explains the relationship between the variables is linear in the parameters, i.e. $f(x|\theta)=\sum_{p=1}^k \theta_p g_p(x)$, the problem is more tractable (see 8.2). If not, we are dealing with non-linear regression (8.7)

2. Dimensionality: the number of independent variables can become a limiting factor

3. Error behavior: from significant, non gaussian errors in both dependent and independent variables (no ready-to-use general method) to negligible errors in both y and x

## 8.2 Linear Models

Linear models are those with linear dependencies on the model parameters:

$f(x|\theta)=\sum_{p=1}^k \theta_p g_p(x)$

where the functions $g_p(x)$ do not need to be linear on the independent variables

### 8.2.1 Multivariate Regression

If the functions $g_p(x)$ are indeed linear, we are dealing with linear regression. If we also consider that the error in the dependent variables is negligible, then

$y_i=\theta_0+\theta_1 x_{i1}+\theta_2x_{i2}+...+\theta_kx_{ik}+\epsilon_i$

and $\epsilon_i$ is the error in the dependent variable. Here we have data for $(y_i, \,x_{ip})$, and want to solve for the k parameters $\theta_p$. As we saw in the first day of class, we have more equations (N) than unkowns (k), and therefore usually there will not be a solution that passes through all the data points. The alternative is to find a set of parameters that allow us to be closely enough to the data points. This requires the definition of a norm (distance) so that we can minimize it. In general this norm is L2 and the method is then that of least squares.

If the erros on the dependent variable are gaussian ($\epsilon_i=\mathcal{N}(0,\sigma_i$)), there is a solution that maximizes the likelihood that is easy to compute

Rewriting the problem in matrix form we have:

$Y = M\theta$ 

and the parameters are given by

$\theta = \left( M^TC^{-1}M \right)^{-1} \left(M^TC^{-1}Y\right)$

with the uncertainties 

$\Sigma_\theta = \left( M^TC^{-1}M \right) ^{-1}$

This solution minimizes the sum of the square errors, given by 

$\left(Y- \theta M \right)^TC^{-1}\left(Y- \theta M \right)$

and is valid even if the covariance matrix for the dependent variables, C is not diagonal (i.e. there is correlation between the measures of the dependent variable).

If $M^TM$ is inversible, then there are enough degrees of freedom (datapoints) to constrain the model

### 8.2.2 Polynomial and Basis Function Regression

In many instances, a linear model with linear functions of the dependent variables is not adequate to describe the data. In this case we can use different functions $g_p(x)$, while keeping the linearity of the model on the parameters.

The problem is the same as the one described in 8.2.1, since we are just transforming a set of independent variables through a function

$(x_{i1},\, x_{i2},\, ... \,,x_{ik}) \to (g_1(x_i),\,g_2(x_i),\,...,g_k(x_i))$


Some of the more popular functions used are polynomials (in part due to the fact that almost any function can be Taylor expanded in polynomials) and gaussians. There are routines in astroML that make easy the implementation of these and we are going to see it with an example

### Example: Supernovas

We are going to follow the book's example, in which it generates the distance modulus of SNIa assuming a cosmological model, and we are interested in estimating the parameters of the model that best fit the data

$\mu(z) = \left((1+z)\frac{c}{H_0}\int \frac{dz}{(\Omega_m(1+z)^3+\Omega_\Lambda)^{1/2}}  \right)$

The real model is clearly non linear, but if we do not know that before hand, we may try to fit a linear model

```python
# The usual imports
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
# Routines to perform the regression
from astroML.linear_model import LinearRegression, PolynomialRegression, BasisFunctionRegression
# Functions to generate fake data
from astroML.cosmology import Cosmology
from astroML.datasets import generate_mu_z
```

```python
# Generate the "fake" data
points = 100
z_sample, mu_sample, dmu = generate_mu_z(100, random_state=0)
# And this is the model that the noisy data follow in reality (just for comparison)
cosmo = Cosmology()
z = np.linspace(0.01, 2, 1000)
mu_true = np.asarray(map(cosmo.mu, z))
```

```python
# Plots the data
plt.errorbar(z_sample, mu_sample, dmu, fmt='.b', ecolor='b', lw=1)
plt.xlabel('$z$')
plt.ylabel('$\mu$')
plt.plot(z, mu_true, '--', c='gray')
```

```python
# Linear regression
model_1 = LinearRegression()
model_1.fit(z_sample[:,None], mu_sample, dmu)
mu_lin = model_1.predict(z[:,None])
mu_lin_fit = model_1.predict(z_sample[:,None])
res_lin = mu_lin_fit - mu_sample
# Polynomial
n = 6
model_2 = PolynomialRegression(n)
model_2.fit(z_sample[:,None], mu_sample, dmu)
mu_pol = model_2.predict(z[:,None])
mu_pol_fit = model_2.predict(z_sample[:,None])
res_pol = mu_pol_fit - mu_sample
# Basis function (gaussian)
basis_mu = np.linspace(0, 2, 15)[:, None]
basis_sigma = 3 * (basis_mu[1] - basis_mu[0])
model_3 = BasisFunctionRegression('gaussian', mu = basis_mu, sigma = basis_sigma)
model_3.fit(z_sample[:,None], mu_sample, dmu)
mu_gauss = model_3.predict(z[:,None])
mu_gauss_fit = model_3.predict(z_sample[:,None])
res_gauss = mu_gauss_fit - mu_sample
```

```python
# Plot the results
f, ax = plt.subplots(1,3,sharex=False,sharey=False, figsize=(20,5))
ax[0].errorbar(z_sample, mu_sample, dmu, fmt='.b', ecolor='b', lw=1)
ax[0].plot(z, mu_true, '--', c='gray')
ax[0].plot(z, mu_lin, '-k')
ax[0].set_xlabel('$z$')
ax[0].set_ylabel('$\mu$')

ax[1].errorbar(z_sample, mu_sample, dmu, fmt='.b', ecolor='b', lw=1)
ax[1].plot(z, mu_true, '--', c='gray')
ax[1].plot(z, mu_pol, '-k')
ax[1].set_xlabel('$z$')
ax[1].set_ylabel('$\mu$')
ax[1].set_ylim(32,50)

ax[2].errorbar(z_sample, mu_sample, dmu, fmt='.b', ecolor='b', lw=1)
ax[2].plot(z, mu_true, '--', c='gray')
ax[2].plot(z, mu_gauss, '-k')
ax[2].set_xlabel('$z$')
ax[2].set_ylabel('$\mu$')
ax[2].set_ylim(32,50)
```

### Summary: underlying hypothesis when applying linear regression

1. The errors in the independent variables are negligible (small compared with the scatter)

2. The errors in the dependent variable are gaussian, although they can be different for each datapoint (heteroscedastic) and they can be correlated (non diagonal covariance matrix)

```python
# Plot the residuals
f, ax = plt.subplots(1,3,sharex=False,sharey=False, figsize=(20,5))
ax[0].plot(z_sample, res_lin, '.b')
ax[0].plot([0,1.8],[0,0])
ax[0].set_xlabel('$z$')
ax[0].set_ylabel('$\hat{\mu}-\mu$')

ax[1].plot(z_sample, res_pol, '.b')
ax[1].plot([0,1.8],[0,0])
ax[1].set_xlabel('$z$')
ax[1].set_ylabel('$\hat{\mu}-\mu$')
ax[1].set_ylim(-3,3)

ax[2].plot(z_sample, res_gauss, '.b')
ax[2].plot([0,1.8],[0,0])
ax[2].set_xlabel('$z$')
ax[2].set_ylabel('$\hat{\mu}-\mu$')
ax[2].set_ylim(-3,3)
```

In the leftmost plot there could be some structure in the residuals, indication that using non linear functions for $x_i$ may be a good idea

```python
from scipy.stats import probplot, shapiro
f, ax = plt.subplots(1,3,sharex=False,sharey=False, figsize=(20,5))
plt.subplot(1,3,1)
a=probplot(res_lin, plot=plt)
plt.subplot(1,3,2)
b=probplot(res_pol, plot=plt)
plt.subplot(1,3,3)
c=probplot(res_gauss, plot=plt)
```

### Not to forget when applying linear regression

1.  The fitted model only makes sense within the space defined by the data (beware of extrapolating without realizing!)

2. Using polynomials or basis functions may improve the model, but be careful not to overfit the data

3. The underlying model may not be linear: think about how to interpret the parameters obtained
